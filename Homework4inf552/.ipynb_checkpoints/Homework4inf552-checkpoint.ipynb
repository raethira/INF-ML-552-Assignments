<h1>1. Active Learning Using Support Vector Machines<h1>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

<h2>(a) Importing data</h2>


```python
#Load data
from sklearn.utils import shuffle

datapluslabels = np.loadtxt("Dataset/data_banknote_authentication.txt",delimiter=",")
#datapluslabels= shuffle(datapluslabels, random_state=0)

data = datapluslabels[:,:4]
labels = datapluslabels[:,4]
print ("Data dimensions : ",np.shape(data))
print ("Labels          : ",np.shape(labels))

#for i in range(3):
#print ("Data = ",data[i]," Label = ",labels[i]) 
col = np.where(labels==0,'k',np.where(labels==1,'b','r'))
#print(col)
```

<h2>Train test split</h2>

```python
#Splitting data into training and test sets

testset = []
trainingset = []
targetset= []
zerocounter, onecounter = 0,0
for i in range(np.shape(datapluslabels)[0]):
    if datapluslabels[i][4] == 0:
        zerocounter += 1
        if zerocounter <= 312:
            testset.append(datapluslabels[i])
        else:
            trainingset.append(datapluslabels[i])
            targetset.append(datapluslabels[i][4])
    elif datapluslabels[i][4] == 1:
        onecounter += 1
        if onecounter <= 160:
            testset.append(datapluslabels[i])
        else:
            trainingset.append(datapluslabels[i])
            targetset.append(datapluslabels[i][4])

test = np.array(testset)
train = np.array(trainingset)
target = np.array(targetset)
#print(test)

'''

Selected the first 236 rows of Class 0 and the first 236 rows of Class 1 as the test set and the rest of the data as the training set, fromt the shuffled data.

'''
print ("Test data shape      : ",test.shape)
print ("Training data shape  : ",train.shape)
#print ("Targetset data shape : ",target.shape)
#print (test[0][:4])
#print(test[0][:4].reshape(1,-1))
#print(np.transpose(testset)[4])
#print(train.shape[0])
#print(test)
#print((test[:][4]))

#print(train[449,:])
#print(train[450,:])
```

<h2>(b).i) Passive learning </h2>

```python
cols=['variance','skewness','curtosis','entropy']
cols2=['variance','skewness','curtosis','entropy','class']

X_train=pd.DataFrame(train[:,:4],columns=cols)
#print(X_train)
X_test=pd.DataFrame(test[:,:4],columns=cols)

y_train=pd.DataFrame(train[:,4],columns=['class'])
y_test=pd.DataFrame(test[:,4],columns=['class'])
#print(y_test)
train_df_master=pd.DataFrame(train,columns=cols2)
test_df_master=pd.DataFrame(test,columns=cols2)

#print(train_df_master.shape)
#print(test_df_master.shape)
#print(train_df_master.iloc[449,:])
#print(train_df_master.iloc[450,:])
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import LinearSVC
from sklearn.model_selection import KFold



kf=KFold(n_splits=10)
svc=LinearSVC(penalty='l1',dual=False)

#parameters={'C':np.linspace(0.01,5,50)}
parameters={'C':[1,10]}
gsc=GridSearchCV(svc,parameters,cv=kf)

passiveErrors = np.zeros(90)
errors=[]

for n in range(0,50):
    temp_DF=pd.DataFrame()
    errors=[]
    train_df_parent=train_df_master
    test_df_parent=test_df_master
    x_axis=[]
    
    for i in range(0,90):

        #print(i)
        tempArray_0=train_df_parent.loc[train_df_parent['class'] == 0].sample(n=5)
        tempArray_1=train_df_parent.loc[train_df_parent['class'] == 1].sample(n=5)
        x=[temp_DF,tempArray_0,tempArray_1]
        temp_DF=pd.concat(x)

        #print(temp_DF.shape)
        train_df_parent=train_df_parent[~train_df_parent.index.isin(temp_DF.index)]
        #print(train_df_parent.shape)

        gsc.fit(temp_DF.loc[:,:'entropy'],temp_DF['class'])
        accuracy_score=gsc.score(test_df_parent.loc[:,:'entropy'],test_df_parent['class'])
        err=1-accuracy_score
        errors.append(err)
        passiveErrors[i]+=err
        x_axis.append(i)
#print(errors)
        
```

```python
print("\n")

plt.title('Test-errors for 90 SVMs- Passive Learning at 50th step')
plt.xlabel('Number of SVMs')
plt.ylabel('Test Error')
plt.plot(x_axis,errors)
plt.show()
#print(errors)
```

```python
print(np.array(passiveErrors).shape)
```

<h2>ii.) Active Learning</h2>

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import LinearSVC
from sklearn.model_selection import KFold


kf=KFold(n_splits=10)
svc=LinearSVC(penalty='l1',dual=False)

#parameters={'C':np.linspace(0.01,5,50)}
parameters={'C':[1,10]}
gsc=GridSearchCV(svc,parameters,cv=kf)

activeErrors = np.zeros(90)
errors=[]

for n in range(0,50):
    temp_DF=pd.DataFrame()
    errors=[]
    train_df_parent=train_df_master
    test_df_parent=test_df_master
    x_axis=[]
    
    tempArray_0=train_df_parent.loc[train_df_parent['class'] == 0].sample(n=5)
    tempArray_1=train_df_parent.loc[train_df_parent['class'] == 1].sample(n=5)
        
    
    for i in range(0,90):

        #print(i)
        x=[temp_DF,tempArray_0,tempArray_1]
        temp_DF=pd.concat(x)

        #print(temp_DF.shape)
        train_df_parent=train_df_parent[~train_df_parent.index.isin(temp_DF.index)]
        #print(train_df_parent.shape)
        
        gsc.fit(temp_DF.loc[:,:'entropy'],temp_DF['class'])
        accuracy_score=gsc.score(test_df_parent.loc[:,:'entropy'],test_df_parent['class'])
        err=1-accuracy_score
        errors.append(err)
        activeErrors[i]+=err
        x_axis.append(i)
        
        if i==89:
            continue
        y=gsc.decision_function(train_df_parent.loc[:,:'entropy'])
        y=abs(y)
        x_loc=y.argsort()[:10]
       
        tempArray_0=train_df_parent.iloc[x_loc,:]
        tempArray_1=pd.DataFrame()
        
        
#print(errors)
#x_axis.append(89)    
```

```python
print("\n")

plt.title('Test-errors for 90 SVMs- Active Learning at 50th step')
plt.xlabel('Number of SVMs')
plt.ylabel('Test Error')
plt.plot(x_axis,errors)
plt.show()
#print(errors)
```

```python
print(np.array(activeErrors).shape)
```

<h2>(c). Monte Carlo simulation </h2>

```python
pl = [p/50 for p in passiveErrors]
al = [a/50 for a in activeErrors]

plt.title('Average Test-errors for 90 SVMs- Passive vs Active Learning')
plt.xlabel('Number of SVMs')
plt.ylabel('Test Error')
plt.plot(x_axis, al,'r',label="Active learning")
plt.plot(x_axis, pl,'b',label="Passive learning")
plt.legend()
plt.show()

```

<pre>
The error rate decreases rapidly for both, while adding more data points.  Active learning performs better at earlier stages itself.  The smoothness in the curve without much kinks denotes a more stable model, even when the n_samples are less.

<h2>(b) i.) Exact match and Hamming loss</h2>

<pre>

Exact match:
=============

In multilabel classification, this function computes subset accuracy: the set of labels predicted for a sample must exactly match the corresponding set of labels in y_true. Exact match is the most strict metric, indicating the percentage of samples that have all their labels classified correctly.

In multilabel classification, the zero_one_loss function corresponds to the subset zero-one loss: for each sample, the entire set of labels must be correctly predicted, otherwise the loss for that sample is equal to one.

The zero_one_loss function computes the sum or the average of the 0-1 classification loss (L_{0-1}) over n_{\text{samples}}. By default, the function normalizes over the sample. To get the sum of the L_{0-1}, set normalize to False.

In multilabel classification, the zero_one_loss scores a subset as one if its labels strictly match the predictions, and as a zero if there are any errors. By default, the function returns the percentage of imperfectly predicted subsets. To get the count of such subsets instead, set normalize to False

If \hat{y}_i is the predicted value of the i-th sample and y_i is the corresponding true value, then the 0-1 loss L_{0-1} is defined as:

L_{0-1}(y_i, \hat{y}_i) = 1(\hat{y}_i \not= y_i)

where 1(x) is the indicator function.


Hamming score/ loss:
=====================

The Hamming loss is the fraction of labels that are incorrectly predicted.It is the fraction of the wrong labels to the total number of labels.

The hamming_loss computes the average Hamming loss or Hamming distance between two sets of samples.

If \hat{y}_j is the predicted value for the j-th label of a given sample, y_j is the corresponding true value, and n_\text{labels} is the number of classes or labels, then the Hamming loss L_{Hamming} between two samples is defined as:

L_{Hamming}(y, \hat{y}) = \frac{1}{n_\text{labels}} \sum_{j=0}^{n_\text{labels} - 1} 1(\hat{y}_j \not= y_j)

where 1(x) is the indicator function.

</pre>

<h2>i.) Hamming loss and exact match for 'Family'<h2>

```python
from sklearn.metrics import hamming_loss
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

svc=SVC()
svc.fit(X_train,y_train["Family"])
preds=svc.predict(X_test)
h=hamming_loss(y_test["Family"],preds)
print("Hamming loss : ",h)
e=accuracy_score(y_test["Family"], preds)
print("Exact match  : ",e)
```

<h2> Hamming loss and exact match for 'Species'<h2>

```python
from sklearn.metrics import hamming_loss
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

svc=SVC()
svc.fit(X_train,y_train["Species"])
preds=svc.predict(X_test)
h=hamming_loss(y_test["Species"],preds)
print("Hamming loss : ",h)
e=accuracy_score(y_test["Species"], preds)
print("Exact match  : ",e)
```

<h2> Hamming loss and exact match for 'Genus'<h2>

```python
from sklearn.metrics import hamming_loss
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

svc=SVC()
svc.fit(X_train,y_train["Genus"])
preds=svc.predict(X_test)
h=hamming_loss(y_test["Genus"],preds)
print("Hamming loss : ",h)
e=accuracy_score(y_test["Genus"], preds)
print("Exact match  : ",e)
```

<h2>ii.) Gaussian kernel-SVM <h2>

```python
from sklearn.multiclass import OneVsRestClassifier
from sklearn.model_selection import RandomizedSearchCV

kf = KFold(10)
C=[0.1,0.01,1,10,100,1000,10000]
gamma = [0.00001,0.0001,0.001,0.01,0.1,0.5,1,10,100]

parameters = {'estimator__C':C,'estimator__gamma':gamma}
#parameters = {'C':C,'gamma':gamma}

print("Label: Genus")
print('============')

model=OneVsRestClassifier(SVC(tol=0.1))
gsc = GridSearchCV(model, parameters, cv=kf,refit=True, n_jobs=5)  # to fasten the convergence

gsc.fit(X_train, y_train["Genus"])

print("Weight of the SVM penalty : ", gsc.best_params_)
print("Best Score                : ", gsc.best_score_)
width=1/(np.sqrt(2*gsc.best_params_.get('estimator__gamma')))
print("Width of Gaussian Kernel  : ",width)

preds_g = gsc.predict(X_test)
score = accuracy_score(y_test["Genus"], preds_g)

print("Exact match               : ", score)
h_g=hamming_loss(y_test["Genus"],preds_g)
print('Hamming Loss              : ',h_g)
```

```python
print("Label: Family")
print('============')

gsc.fit(X_train, y_train["Family"])

print("Weight of the SVM penalty : ", gsc.best_params_)
print("Best Score                : ", gsc.best_score_)
width=1/(np.sqrt(2*gsc.best_params_.get('estimator__gamma')))
print("Width of Gaussian Kernel  : ",width)

preds_f = gsc.predict(X_test)
score = accuracy_score(y_test["Family"], preds_f)

print("Exact match               : ", score)
h_f=hamming_loss(y_test["Family"],preds_f)
print('Hamming Loss              : ',h_f)

```

```python
print("Label: Species")
print('============')

gsc.fit(X_train, y_train["Species"])

print("Weight of the SVM penalty : ", gsc.best_params_)
print("Best Score                : ", gsc.best_score_)
width=1/(np.sqrt(2*gsc.best_params_.get('estimator__gamma')))
print("Width of Gaussian Kernel  : ",width)

preds_s = gsc.predict(X_test)
score = accuracy_score(y_test["Species"], preds_s)

print("Exact match               : ", score)
h_s=hamming_loss(y_test["Species"],preds_s)
print('Hamming Loss              : ',h_s)

```

```python
correct=0
wrong=0
#print(y_test["Family"].iloc[0])
for i in range(0,len(y_test["Species"])):
    if((preds_f[i]==y_test["Family"].iloc[i]) and (preds_s[i]==y_test["Species"].iloc[i]) and (preds_g[i]==y_test["Genus"].iloc[i])):
        correct+=1
    else:
        wrong+=1

avg_exact=wrong/len(y_test["Species"])
avg_hamm=(h_s+h_f+h_g)/3

print("Average Hamming Loss     : ",avg_hamm)
print("Average exact match loss : ",avg_exact)
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

X_train, X_test, y_train, y_test = train_test_split(df.iloc[:,:22], df.iloc[:,22:], test_size=0)

#print(X_train.shape)
#print(y_train.shape)

clusters=np.arange(2,20)
results=[]
for size in clusters:
    model=KMeans(n_clusters=size).fit(X_train)
    preds=model.predict(X_train)
    results.append(silhouette_score(X_train,preds))
best_K=np.array(results).argmax()
#print(results)
print('Best K : ',best_K+2)

hf=[]
hs=[]
hg=[]

for i in range(0,50):

    kmeans=KMeans(n_clusters=4).fit(X_train)
    clusters=kmeans.labels_
    #print(clusters)

    i0 = np.argwhere(clusters == 0)
    i1 = np.argwhere(clusters == 1)
    i2 = np.argwhere(clusters == 2)
    i3 = np.argwhere(clusters == 3)

    #print(y_train['Family'].shape)
    #print(i0.ravel())



    #print("Family :")
    #print('=======')

    f1=y_train['Family'].iloc[i0.ravel()]
    #print("Cluster1 : ")
    #print(f1.value_counts())
    #print('\n')
    f2=y_train['Family'].iloc[i1.ravel()]
    #print("Cluster2  : ")
    #print(f2.value_counts())
    #print('\n')
    f3=y_train['Family'].iloc[i2.ravel()]
    #print("Cluster3 : ")
    #print(f3.value_counts())
    #print('\n')
    f4=y_train['Family'].iloc[i3.ravel()]
    #print("Cluster4 : ")
    #print(f4.value_counts())


    #print("\n")

    #print("Species :")
    #print('=======')

    s1=y_train['Species'].iloc[i0.ravel()]
    #print("Cluster1 : ")
    #print(s1.value_counts())
    #print('\n')
    s2=y_train['Species'].iloc[i1.ravel()]
    #print("Cluster2  : ")
    #print(s2.value_counts())
    #print('\n')
    s3=y_train['Species'].iloc[i2.ravel()]
    #print("Cluster3 : ")
    #print(s3.value_counts())
    #print('\n')
    s4=y_train['Species'].iloc[i3.ravel()]
    #print("Cluster4 : ")
    #print(s4.value_counts())


    #print("\n")
    #print("GENUS :")
    #print('======')

    g1=y_train['Genus'].iloc[i0.ravel()]
  #  print("Cluster1 : ")
   # print(s1.value_counts())
    #print('\n')
    g2=y_train['Genus'].iloc[i1.ravel()]
    #print("Cluster2  : ")
    #print(s2.value_counts())
    #print('\n')
    g3=y_train['Genus'].iloc[i2.ravel()]
    #print("Cluster3 : ")
    #print(s3.value_counts())
    #print('\n')
    g4=y_train['Genus'].iloc[i3.ravel()]
    #print("Cluster4 : ")
    #print(s4.value_counts())


    from sklearn.metrics import hamming_loss

    pred=kmeans.predict(X_train)


    #print(y_train['Family'])

    family=pd.DataFrame(pred,columns=['Family'])
    #print(new_family)
    genus=pd.DataFrame(pred,columns=['Genus'])
    species=pd.DataFrame(pred,columns=['Species'])

    family=family.replace((0,1,2,3),(f1.value_counts().index.values[0],f2.value_counts().index.values[0],f3.value_counts().index.values[0],f4.value_counts().index.values[0]))
    species=species.replace((0,1,2,3),(s1.value_counts().index.values[0],s2.value_counts().index.values[0],s3.value_counts().index.values[0],s4.value_counts().index.values[0]))
    genus=genus.replace((0,1,2,3),(g1.value_counts().index.values[0],g2.value_counts().index.values[0],g3.value_counts().index.values[0],g4.value_counts().index.values[0]))


    hf.append(hamming_loss(y_train['Family'],family))
    hg.append(hamming_loss(y_train['Genus'],genus))
    hs.append(hamming_loss(y_train['Species'],species))




```
