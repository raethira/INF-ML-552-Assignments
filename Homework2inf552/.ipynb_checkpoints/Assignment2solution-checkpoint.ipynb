```python
def compute_features(colvalues, prepend_col_name):
    return_dict = {}
    return_dict.update({'{}_mean'.format(prepend_col_name):mean(colvalues)})
    return_dict.update({'{}_standard_deviation'.format(prepend_col_name):std(colvalues)})
    return_dict.update({'{}_skew'.format(prepend_col_name):skew(colvalues)})
    return_dict.update({'{}_kurtosis'.format(prepend_col_name):kurtosis(colvalues, fisher=False)})
    return_dict.update({'{}_min'.format(prepend_col_name):min(colvalues)})
    return_dict.update({'{}_max'.format(prepend_col_name):max(colvalues)})
    return return_dict
```

```python
def find_features(temp_df, label, which_df) -> 'modifies the global temp_(train|test)_features_df dataframe => finds features from the input temp_df':
    global train_features_df
    global test_features_df
    dataframe_dictionary = {}
    for column in temp_df.columns.drop(["file","label","time"]):
        dataframe_dictionary.update(compute_features(list(temp_df[column].values), prepend_col_name=column))
    data = list(dataframe_dictionary.values())
    data.append(label)
    keys = list(dataframe_dictionary.keys())
    keys.append("label")
    if which_df == "train":
        if train_features_df.empty:
            train_features_df = pd.DataFrame(np.array(data).reshape(1,37), columns=keys)
            train_features_df.iloc[0] = pd.to_numeric(train_features_df.iloc[0], errors="coerce")
            train_features_df = train_features_df.round(4)
            train_features_df["label"][0] = label
        else:
            new_series = pd.Series(data, index=train_features_df.columns)
            train_features_df = train_features_df.append(new_series, ignore_index=True)
    else:
        if test_features_df.empty:
            test_features_df = pd.DataFrame(np.array(data).reshape(1,37), columns=keys)
            test_features_df.iloc[0] = pd.to_numeric(test_features_df.iloc[0], errors="coerce")
            test_features_df = test_features_df.round(4)
            test_features_df["label"][0] = label
        else:
            new_series = pd.Series(data, index=test_features_df.columns)
            test_features_df = test_features_df.append(new_series, ignore_index=True)
```

```python
def find_temp_features(temp_df, label, which_df) -> 'modifies the global temp_(train|test)_features_df dataframe => finds features from the input temp_df':
    global temp_train_features_df
    global temp_test_features_df
    dataframe_dictionary = {}
    for column in temp_df.columns.drop(["file","label","time"]):
        dataframe_dictionary.update(compute_features(list(temp_df[column].values), prepend_col_name=column))
    data = list(dataframe_dictionary.values())
    data.append(label)
    keys = list(dataframe_dictionary.keys())
    keys.append("label")
    if which_df == "train":
        if temp_train_features_df.empty:
            temp_train_features_df = pd.DataFrame(np.array(data).reshape(1,37), columns=keys)
            temp_train_features_df.iloc[0] = pd.to_numeric(temp_train_features_df.iloc[0], errors="coerce")
            temp_train_features_df = temp_train_features_df.round(4)
            temp_train_features_df["label"][0] = label
        else:
            new_series = pd.Series(data, index=temp_train_features_df.columns)
            temp_train_features_df = temp_train_features_df.append(new_series, ignore_index=True)
    else:
        if temp_test_features_df.empty:
            temp_test_features_df = pd.DataFrame(np.array(data).reshape(1,37), columns=keys)
            temp_test_features_df.iloc[0] = pd.to_numeric(temp_test_features_df.iloc[0], errors="coerce")
            temp_test_features_df = temp_test_features_df.round(4)
            temp_test_features_df["label"][0] = label
        else:
            new_series = pd.Series(data, index=temp_test_features_df.columns)
            temp_test_features_df = temp_test_features_df.append(new_series, ignore_index=True)
```

```python
def equipartition_finder(length, no_of_partitions) -> 'returns list of lists eg. for 480/2 => [[0,239], [240,479]]':
    if no_of_partitions == 1:
        return [[0, length-1]]
    equipartition_mode_length = length//no_of_partitions # here the mode refers to the most common length for example 480/7 will 6-68's and 1-72 partition
    t = 0
    partitions = []
    for i in range(no_of_partitions-1):
        partitions.append([t, t+equipartition_mode_length-1])
        t = t+equipartition_mode_length
    partitions.append([t, length-1])
    return partitions
```

```python
avg_rmse = []
for i in range(2,11):
    temp_train_features_df = pd.DataFrame()
    temp_test_features_df = pd.DataFrame()
    for pair in train_file_label_pair:
        temp_df = train_df[train_df["file"]==pair[0]]
        temp_df = temp_df[temp_df["label"]==pair[1]]
        equipartitions = equipartition_finder(len(temp_df), i)
        count = 0
        for partition in equipartitions:
            find_temp_features(temp_df.iloc[partition[0]:partition[1]], label=pair[1], which_df="train")
    temp_train_features_df = temp_train_features_df[final_features_df.columns]
    temp_train_features_df['label'] = temp_train_features_df['label'].apply(lambda x: 1.0 if 'bending' in x else 0.0)    
    temp_train_features_df = (temp_train_features_df - temp_train_features_df.min())/(temp_train_features_df.max() - temp_train_features_df.min())
    for pair in test_file_label_pair:
        temp_df = test_df[test_df["file"]==pair[0]]
        temp_df = temp_df[temp_df["label"]==pair[1]]
        equipartitions = equipartition_finder(len(temp_df), i)
        count = 0
        for partition in equipartitions:
            find_temp_features(temp_df.iloc[partition[0]:partition[1]], label=pair[1], which_df="test")
    temp_test_features_df = temp_test_features_df[final_features_df.columns]    
    kf = KFold(n_splits = 5, shuffle=True)
    X = temp_train_features_df[temp_train_features_df.columns.drop(['label'])]
    y = temp_train_features_df['label']        
    avg_of_all_rmses = 0
    for train_index, test_index in kf.split(temp_train_features_df):
        X_train, X_test = X.loc[train_index], X.loc[test_index]
        y_train, y_test = y.loc[train_index], y.loc[test_index]
        X_train = add_constant(X_train)
        logit = Logit(y_train, X_train.astype(float))
        model = logit.fit(method='bfgs', )
        print(model.summary())
        try:
            avg_of_all_rmses = avg_of_all_rmses + model.mse_total
        except:
            pass
    avg_of_all_rmses = avg_of_all_rmses / 5
    avg_rmse.append(avg_of_all_rmses)
    break
        
print(avg_rmse)
```

### Right and wrong way to perform cross validation
#### Right way
- There are two right ways for this step
    - First correct method
        - Cross validate to find the best value of L (data augmentation factor)
        - Then on all the instances obtained for that value of L use p-values or backward selection to get the most significant features
    - Second correct method
        - Find the best features and the best value of L simultaneously
        - This implies that for every value of L the features obtained after cross validation are different
        - Since L1 and L2 penalization does this too this method is accepted as the right way
        
#### Wrong way
- Obtain the best features from the model based on p-values or backward selection 
- Then find the best value of L using cross validation

```python
def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    fig = plt.figure(figsize=(20,10))
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        pass
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
```
