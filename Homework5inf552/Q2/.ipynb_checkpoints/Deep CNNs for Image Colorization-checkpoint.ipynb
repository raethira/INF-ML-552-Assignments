<h1>2. Deep CNNs for Image Colorization</h2>

<h2>(a) Downolading the dataset</h2>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import tflearn
from tflearn.data_utils import shuffle, to_categorical
from tflearn.layers.core import input_data, dropout, fully_connected
from tflearn.layers.conv import conv_2d, max_pool_2d
from tflearn.layers.estimator import regression
from tflearn.data_preprocessing import ImagePreprocessing
from tflearn.data_augmentation import ImageAugmentation
from tflearn.datasets import cifar10

# Referring https://github.com/tflearn/tflearn/blob/master/examples/images/convnet_cifar10.py
# pip3 install tflearn
# https://github.com/tflearn/tflearn/issues/428

# Data loading and preprocessing
(X_train, y_train), (X_test, y_test) = cifar10.load_data()

y_train = to_categorical(y_train, nb_classes=10)
y_train = y_train[:,2]
y_test = to_categorical(y_test, nb_classes=10)
y_test = y_test[:,2]

print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)

```

<h2>(b) Extract the class birds</h2>

```python
birds = np.argwhere(y_train==1)
X2_train = X_train[birds.ravel()]

test = np.argwhere(y_test==1)
X2_test = X_test[test.ravel()]

print(X2_train.shape)
print(y_train.shape)
print(X2_test.shape)
print(y_test.shape)
```

```python
plt.imshow(X2_train[0])
plt.show()
plt.imshow(X2_train[1])
plt.show()
plt.imshow(X2_train[2])
plt.show()
```

```python
print('Training set : ', X2_train.shape)
print('Test set     : ', X2_test.shape)
```

<h2>(c) Choosing most of the pixels</h2>

```python
X1_train = X2_train.reshape(5120000,3)     # 5000*32*32=5120000
X1_test = X2_test.reshape(1024000,3)     # 1000*32*32=1024000

print('Training set : ', X1_train.shape)
print('Test set     : ', X1_test.shape)
```

<h2>(d) K-means clustering</h2>

```python
from sklearn.cluster import KMeans

km1 = KMeans(n_clusters=4,max_iter=1000,random_state=0)
km1.fit(X1_train)
center1 = km1.cluster_centers_
```

```python
center_train = center1
print('Training centers : ',center_train)
```

```python
train_color_label = km1.predict(X1_train) 
X_train_color = []
counter = 0
for pixel in X1_train:
    X_train_color.append(center_train[train_color_label[counter]])
    counter += 1
```

```python
km2 = KMeans(n_clusters=4,max_iter=1000,random_state=0)
X_test_color = km2.fit_transform(X1_test)
center2 = km2.cluster_centers_
```

```python
center_test = center2
print('Test centers : ',center_test)
```

```python
test_color_label = km2.predict(X1_test) 
X_test_color = []
counter = 0
for pixel in X1_test:
    X_test_color.append(center_test[test_color_label[counter]])
    counter += 1
```

```python
v_X_train_color = np.reshape(X_train_color,(5000,32,32,3)) 
v_X_test_color = np.reshape(X_test_color,(1000,32,32,3)) 
```

```python
plt.imshow(v_X_train_color[2])
plt.show()
plt.imshow(v_X_test_color[1])
plt.show()
```

<h2>(e) RGB to Grayscale</h2>

```python

from skimage.color import rgb2gray

X_train_gray = rgb2gray(X2_train)
X_test_gray = rgb2gray(X2_test)

print (X_train_gray.shape)
plt.imshow(X_train_gray[0],cmap='gray')
plt.show()

print (X_test_gray.shape)
plt.imshow(X_test_gray[0],cmap='gray')
plt.show()
```

<h2>(f) Deep CNN</h2>

```python
import numpy as np
import keras
from keras.models import Sequential
from keras.layers import Dense, Flatten
from keras.layers import Conv2D, MaxPooling2D, Reshape
from keras.callbacks import ModelCheckpoint
```

```python
# Model definition

model = Sequential()
# Convolution layer with 16 filters, 5x5 filter size, stride = 1 and relu activation
model.add(Conv2D(16,strides=1, kernel_size=(5, 5), input_shape=(32,32,1), padding='same', activation='relu'))
# Pooling layer with pool size 2x2
model.add(MaxPooling2D(pool_size=(2,2),padding='same'))
# Convolution layer with 32 filters, 5x5 filter size, stride = 1 and relu activation
model.add(Conv2D(32,strides=1, kernel_size=(5, 5), padding='same', activation='relu'))
# Pooling layer with pool size 2x2
model.add(MaxPooling2D(pool_size=(2,2),padding='same'))
# Flatten volume
model.add(Flatten())
# ReLU layer 
model.add(Dense((2048),activation='relu'))
# Softmax output layer
model.add(Dense((4096),activation='softmax')) # 32x32x4
model.summary()
# Define optimizer
adam_optimizer = keras.optimizers.Adam(lr=0.001)
# Compile model
model.compile(loss='categorical_crossentropy', optimizer=adam_optimizer)
```

```python
# Define checkpoint
filename = "model_{epoch:02d}_{loss:.4f}.hdf5"
checkpoint = ModelCheckpoint(filename, verbose = 1, save_best_only = False, 
                                       save_weights_only = False, period = 1)
```

```python
# Reshape input data to (32,32,1)

X_train_gray = np.reshape(X_train_gray,(len(X_train_gray),32,32,1))
X_test_gray = np.reshape(X_test_gray,(len(X_test_gray),32,32,1))

print (X_train_gray.shape)
print (X_test_gray.shape)
```

```python
# Create label data to match shape of generated output (32x32xk) where k represents one hot encoding 
# of class of closest center

train_color_label = km1.predict(X1_train) 
print (set(train_color_label))
```

```python
# Can use sklearn OneHotEncoder() but defining manually as k = 4

onehot = {0:[1,0,0,0],1:[0,1,0,0],2:[0,0,1,0],3:[0,0,0,1]}

X_train_ground = []
counter = 0
for pixel in X1_train:
    X_train_ground.append(onehot[train_color_label[counter]])
    counter += 1
print (np.array(X_train_ground).shape)
X_train_ground = np.reshape(X_train_ground,(5000,4096))
print (X_train_ground.shape)
```

```python
test_color_label = km2.predict(X1_test) 

X_test_ground = []
counter = 0
for pixel in X1_test:
    X_test_ground.append(onehot[test_color_label[counter]])
    counter += 1

print (np.array(X_test_ground).shape)
X_test_ground = np.reshape(X_test_ground,(1000,4096))
print (X_test_ground.shape)
```

```python
history = model.fit(X_train_gray, X_train_ground,
                    batch_size=4,
                    epochs=20,
                    verbose=1,
                    shuffle=True,validation_split=0.1,callbacks=[checkpoint])
```
